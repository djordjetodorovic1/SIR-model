```python
import pandas as pd
import matplotlib.pyplot as plt
import os

results_folder = "rezultati1"

for filename in os.listdir(results_folder):
    if filename.endswith(".csv"):
        filepath = os.path.join(results_folder, filename)
        
        df = pd.read_csv(filepath)
        
        plt.figure(figsize=(10,6))
        plt.plot(df['Day'], df['Susceptible'], label='S (Susceptible)', color='blue')
        plt.plot(df['Day'], df['Infected'], label='I (Infected)', color='red')
        plt.plot(df['Day'], df['Recovered'], label='R (Recovered)', color='green')
        
        plt.title(f"SIR Model Simulation - {filename}")
        plt.xlabel("Day")
        plt.ylabel("Number of People")
        plt.legend()
        plt.grid(True)
        
        plot_name = os.path.join(results_folder, filename.replace(".csv", "_plot.png"))
        plt.savefig(plot_name)
        plt.show()
        plt.close()
```
